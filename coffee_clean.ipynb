```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import warnings
warnings.filterwarnings('ignore') 
```

I start by webscraping the coffee consumption data that i found online by doing a request and process the html information with Beautifull Soup

```python
url= "https://cafely.com/blogs/research/which-country-consumes-the-most-coffee?srsltid=AfmBOooy1znaM-b_BvIsZhMLRjZFvsDTyNMtxBeU8Hzi6Y08RJHfbDP5"
```

```python
response= requests.get(url)
```

```python
response.status_code
```

```python
soup= BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```python
html_table = soup.find_all("table", {"id":"full-data-table"})[0]
print(html_table.prettify())
```

By using the method read_html from pandas library I create the df with information collected before

```python
df= pd.read_html(html_table.prettify())[0]
df
```

There are no Nan, duplicates or empty spaces, therefore I can start renaming my columns for better understanding and readability

```python
df.columns= ["Drop", "country", "continent", "consumption_kg", "yearly_coffee_consumption_pcapita_kg", "daily_cofee_consumption_pcapita_cup", "coffe_drinking_year", "lifetime_cup_consumption_cup", "price_per_cup", "total_lifetime_coffee_spending"]
df.head(2)
```

By cleaning my columns, like taking away the symbols, some numeric columns turned as "object" type since i had to convert it to a string in order to use the replace method. Therefore i pass it again as float after the cleaning

```python
df["price_per_cup"]= df["price_per_cup"].str.replace("$", " ")
df["total_lifetime_coffee_spending"]= df["total_lifetime_coffee_spending"].str.replace("$", " ")
df["total_lifetime_coffee_spending"]= df["total_lifetime_coffee_spending"].str.replace(",", ".")
```

```python
df.dtypes
```

```python
df["price_per_cup"]= df["price_per_cup"].astype(float)
df["total_lifetime_coffee_spending"]= df["total_lifetime_coffee_spending"].astype(float)
```

```python
df.dtypes
```

Lastly i drop the unnecessary column, letting only the important columns in the data set

```python
df= df.drop(columns="Drop")
```
